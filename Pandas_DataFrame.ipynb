```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Patient':['Abdul','Rahman','Ajay','Nav','Ibu'],
        'Age':[34,22,28,42,18],
        'Weight':[80,68,72,85,63],
        'Systolic Blood Pressure':[130,110,120,140,100],
        'Diastolic Blood Pressure':[80,60,70,90,50]}
df = pd.DataFrame(data)
print("Original DataFrame: ")
print(df)

df['Systolic Blood Pressure'] = pd.to_numeric(df['Systolic Blood Pressure'])
df['Diastolic Blood Pressure'] = pd.to_numeric(df['Diastolic Blood Pressure'])
df['Height'] = 1.75
df['BMI'] = df['Weight'] / (df['Height']*df['Height'])

print("Modified DataFrame: ")
print(df)

df.plot(kind='bar', x = 'Patient', y = 'BMI', title = 'Barchart for BMI')
plt.xlabel('Patient')
plt.ylabel('BMI')
plt.show()

df.plot.pie(y = 'Systolic Blood Pressure', labels= df['Patient'], autopct = "%1.1f%%", title = ' Systolic Blood Pressure Pie Chart')
plt.show()

df.plot(kind='scatter', x = 'BMI', y = 'Systolic Blood Pressure', title = "BMI vs Systolic Blood Pressure")
plt.show()
```
